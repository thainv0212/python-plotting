# Subplots

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.subplot(1,2,1)
linear_data = np.array([1,2,3,4,5,6,7,8])
plt.plot(linear_data, '-o')
```

```python
exponential_data = linear_data ** 2
plt.subplot(1,2,2)
plt.plot(exponential_data, '-o')
```

```python
plt.subplot(1,2,1)
plt.plot(exponential_data, '-o')
```

```python
plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(linear_data, '-o')
ax2 = plt.subplot(1,2,2, sharey=ax1)
plt.plot(exponential_data, '-x')
```

```python
plt.subplot(3,3,1) == plt.subplot(331)
```

# Histogram

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True)
axs = [ax1, ax2, ax3, ax4]
for n in range(0, len(axs)):
    sample_size = 10 ** (n+1)
    sample = np.random.normal(loc=0.0,scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True)
axs = [ax1, ax2, ax3, ax4]
for n in range(0, len(axs)):
    sample_size = 10 ** (n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample, bins=80)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
plt.figure()
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
plt.scatter(X,Y)
```

```python
import matplotlib.gridspec as gridspec
plt.figure()
gspec = gridspec.GridSpec(3,3)
top_hist = plt.subplot(gspec[0, 1:])
side_hist = plt.subplot(gspec[1:, 0])
lower_hist = plt.subplot(gspec[1:, 1:])
```

```python
Y = np.random.normal(loc=0.0,scale=1.0, size=10000)
Y = np.random.random(size=10000)
lower_hist.scatter(X,Y)
top_hist.hist(X, bins=10000)
s = side_hist.hist(Y, bin=100, orientation='horizontal')
```
