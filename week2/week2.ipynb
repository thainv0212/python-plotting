```python
plt.figure()
observation = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation, x, '-o', observation, y , '-o')
```

### Bar Charts

```python
plt.figure()
linear_data = np.array([0,1,2,3,4,5,6,7])
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3)
```

```python
new_xvals = []
for item in xvals:
    new_xvals.append(item + 0.3)
plt.bar(new_xvals, y, width=0.3, color='red')
```

```python
from random import randint
linear_err= [randint(0,15) for x in range(len(linear_data))]
```

```python
plt.bar(xvals, y, width=0.3, yerr=linear_err)
```

```python
xvals = range(len(linear_data))
plt.figure()
plt.bar(xvals, linear_data, width=0.3, color='b')
plt.bar(xvals, y, width=0.3, bottom=linear_data, color='r')
```

```python
plt.figure()
xvals = range(len(linear_data))
plt.barh(xvals, linear_data, height=0.3, color='b')
plt.barh(xvals, y, height=0.3, left=linear_data, color='r')
```
